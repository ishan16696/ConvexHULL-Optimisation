```python
def scatter(lx,ly,ax,ay,l,w):
    fig = plt.figure(figsize=(l,w))
#     fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.scatter(lx, ly, c='r', marker="s" )
    ax1.scatter(ax,ay, c='b', marker="o")
    plt.show()
```

```python
def read_csv(filename):
    with open(filename, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',')
        x_index = []
        y_index = []
        for row in spamreader:
            x_index.append(float(row[0]))
            y_index.append(float(row[1]))
    return x_index,y_index


def plot_csv(f1,f2,f3,f4):
    x1,y1 = read_csv(f1)
    x2,y2 = read_csv(f2)
    x3,y3 = read_csv(f3)
    x4,y4 = read_csv(f4)
    df = pd.DataFrame({'Jarvis': y1,'monotone': y2,'improved_mono':y3,'graham':y4}, index=x1)
    lines = df.plot(lw = 1, colormap='jet',title='time analysis at different input size of algorithms', figsize=(15,7))
    lines.set_xlabel("Input size in multiple of 1000")
    lines.set_ylabel("Time in milisecond")
def read_txt(fname):
    with open(fname) as f:
        content = f.readlines()
        content = [x.rstrip('\n') for x in content]
        content = content[1:]
        content.pop()
    x_index = []
    y_index = []
    for line in content:
        line = line.split("\t")
        x_index.append(int(line[0]))
        y_index.append(int(line[1]))
    return x_index,y_index
```

## Jarvis March visualization 

```python
lm,ln = read_txt("ishan1/Input/10k.txt")
am,an = read_csv("ishan1/output.csv")
scatter(lm,ln,am,an,20,25)
```

## Modified Quick Hull visualization : 

#### Function to generate quick hull input(input file name) 

```python
lx = []
ly = []
for i in range(30):
    a = random.randint(1,300)
    b = random.randint(1,100)
    if a not in lx:lx.append(a)
    if b not in ly:ly.append(b)
lx[0] = 0
ly[0] = 0
m = min(len(lx) , len(ly))
print(m)
lx = lx[:m]
ly = ly[:m]
lx.append(335)
ly.append(0)

file1 = open("ishan1/Input/qh1.txt","w")
file1.write(str(m+1)+"\n")
for i in range(m+1):
    file1.write(str(lx[i]) + "\t" + str(ly[i]) + "\n")
#write mode
```
