```python
def scatter(lx,ly,ax,ay):
    fig = plt.figure(figsize=(20,25))
#     fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.scatter(lx, ly, c='r', marker="s" )
    ax1.scatter(ax,ay, c='b', marker="o")
    plt.show()
```

```python
def read_csv(filename):
    with open(filename, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',')
        x_index = []
        y_index = []
        for row in spamreader:
            x_index.append(float(row[0]))
            y_index.append(float(row[1]))
    return x_index,y_index


def plot_csv(f1,f2,f3,f4):
    x1,y1 = read_csv(f1)
    x2,y2 = read_csv(f2)
    x3,y3 = read_csv(f3)
    x4,y4 = read_csv(f4)
    df = pd.DataFrame({'Jarvis': y1,'monotone': y2,'improved_mono':y3,'graham':y4}, index=x1)
    lines = df.plot(lw = 1, colormap='jet',title='time analysis at different input size of algorithms', figsize=(15,7))
    lines.set_xlabel("Input size in multiple of 1000")
    lines.set_ylabel("Time in milisecond")
def read_txt(fname):
    with open(fname) as f:
        content = f.readlines()
        content = [x.rstrip('\n') for x in content]
        content = content[1:]
        content.pop()
    x_index = []
    y_index = []
    for line in content:
        line = line.split("\t")
        x_index.append(int(line[0]))
        y_index.append(int(line[1]))
    return x_index,y_index
```
